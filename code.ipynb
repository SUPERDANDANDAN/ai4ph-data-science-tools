```python
data.report_date = pd.to_datetime(data.report_date)
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.cbook as cbook

fig = plt.figure()
ax = plt.axes()

x = data.report_date
y1 = data.total_individuals_at_least_one
#ax.plot(x, y1)
y2 = data.total_individuals_partially_vaccinated
ax.plot(x,y2)
y3 = data.total_individuals_fully_vaccinated 
ax.plot(x,y3)
plt.legend(["parcially vaccinated","fully vaccinated"],loc="lower right")


# Major ticks every 6 months.
fmt_half_year = mdates.MonthLocator(interval=12)
ax.xaxis.set_major_locator(fmt_half_year)

# Minor ticks every month.
fmt_month = mdates.MonthLocator()
ax.xaxis.set_minor_locator(fmt_month)

# Text in the x axis will be displayed in 'YYYY-mm' format.
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))

plt.title("Vaccination status over time")
```
